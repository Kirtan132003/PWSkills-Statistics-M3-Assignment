```python
# Q1) Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and interpret the results.
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

def z_test(sample, pop_mean, pop_std, alpha=0.05):
    """Performs a one-sample Z-test and visualizes the results"""
    sample_mean = np.mean(sample)
    n = len(sample)
    z_score = (sample_mean - pop_mean) / (pop_std / np.sqrt(n))
    p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))  # Two-tailed test
    
    print(f"Sample Mean: {sample_mean}")
    print(f"Z-score: {z_score}")
    print(f"P-value: {p_value}")
    
    if p_value < alpha:
        print("Reject the null hypothesis: The sample mean is significantly different from the population mean.")
    else:
        print("Fail to reject the null hypothesis: No significant difference from the population mean.")
    
    # Visualization
    x = np.linspace(-4, 4, 1000)
    y = stats.norm.pdf(x, 0, 1)
    
    plt.figure(figsize=(8, 5))
    plt.plot(x, y, label="Standard Normal Distribution")
    plt.axvline(z_score, color='r', linestyle='dashed', linewidth=2, label=f'Z-score: {z_score:.2f}')
    plt.fill_between(x, y, where=(x < -abs(z_score)) | (x > abs(z_score)), color='red', alpha=0.3, label='Rejection Region')
    plt.xlabel("Z-score")
    plt.ylabel("Probability Density")
    plt.title("Z-test Visualization")
    plt.legend()
    plt.show()

# Simulate a random sample
test_sample = np.random.normal(loc=50, scale=10, size=30)  # Mean=50, Std=10, Sample size=30

# Perform Z-test
z_test(test_sample, pop_mean=50, pop_std=10)
```

```python
# Q10) Write a Python program to calculate the margin of error for a given confidence level using sample data
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

def calculate_margin_of_error(sample, confidence_level=0.95):
    """Calculates the margin of error for a given confidence level using sample data"""
    sample_mean = np.mean(sample)
    sample_std = np.std(sample, ddof=1)  # Sample standard deviation
    sample_size = len(sample)
    
    # Get the critical z-score
    z_critical = stats.norm.ppf((1 + confidence_level) / 2)
    
    # Calculate margin of error
    margin_of_error = z_critical * (sample_std / np.sqrt(sample_size))
    
    print(f"Sample Mean: {sample_mean}")
    print(f"Margin of Error: {margin_of_error}")
    
    # Visualization
    plt.figure(figsize=(8, 5))
    plt.hist(sample, bins=15, color='blue', alpha=0.7, edgecolor='black')
    plt.axvline(sample_mean, color='red', linestyle='dashed', linewidth=2, label='Sample Mean')
    plt.axvline(sample_mean + margin_of_error, color='green', linestyle='dashed', linewidth=2, label='Upper Bound')
    plt.axvline(sample_mean - margin_of_error, color='orange', linestyle='dashed', linewidth=2, label='Lower Bound')
    plt.xlabel("Sample Values")
    plt.ylabel("Frequency")
    plt.title("Margin of Error Visualization")
    plt.legend()
    plt.show()
    
    return margin_of_error

# Generate sample data
np.random.seed(42)
sample_data = np.random.normal(50, 10, 30)

# Calculate and visualize margin of error
calculate_margin_of_error(sample_data)
```

```python
# Q11) Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process
import numpy as np
import matplotlib.pyplot as plt

def bayes_theorem(prior, likelihood, evidence):
    """Computes the posterior probability using Bayes' Theorem"""
    posterior = (likelihood * prior) / evidence
    return posterior

# Define prior probabilities
prior_A = 0.3  # Probability of hypothesis A being true
prior_B = 0.7  # Probability of hypothesis B being true

# Define likelihoods
likelihood_A = 0.8  # Probability of evidence given hypothesis A
likelihood_B = 0.4  # Probability of evidence given hypothesis B

# Compute total evidence probability (normalization factor)
evidence = (likelihood_A * prior_A) + (likelihood_B * prior_B)

# Compute posterior probabilities
posterior_A = bayes_theorem(prior_A, likelihood_A, evidence)
posterior_B = bayes_theorem(prior_B, likelihood_B, evidence)

# Print results
print(f"Posterior Probability of A: {posterior_A:.4f}")
print(f"Posterior Probability of B: {posterior_B:.4f}")

# Visualization
labels = ['Hypothesis A', 'Hypothesis B']
posteriors = [posterior_A, posterior_B]
plt.bar(labels, posteriors, color=['blue', 'green'])
plt.xlabel("Hypotheses")
plt.ylabel("Posterior Probability")
plt.title("Bayesian Inference: Posterior Probabilities")
plt.ylim(0, 1)
plt.show()
```

```python
# Q19) Write a Python function to check the assumptions (normality, independence, and equal variance) for ANOVAD
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns

# Function to check ANOVA assumptions
def check_anova_assumptions(groups):
    # Normality check using Shapiro-Wilk test
    for i, group in enumerate(groups):
        stat, p = stats.shapiro(group)
        print(f"Shapiro-Wilk Test for Group {i+1}: Statistic={stat:.4f}, P-value={p:.4f}")
        if p < 0.05:
            print(f"Group {i+1} is not normally distributed.")
        else:
            print(f"Group {i+1} is normally distributed.")
    
    # Independence assumption (Usually ensured by study design, no statistical test)
    print("Independence assumption should be ensured during data collection.")
    
    # Equal variance check using Levene's test
    stat, p = stats.levene(*groups)
    print(f"Levene's Test for Equal Variance: Statistic={stat:.4f}, P-value={p:.4f}")
    if p < 0.05:
        print("Variances are not equal across groups.")
    else:
        print("Variances are equal across groups.")

# Generate random samples for three groups
np.random.seed(42)
group1 = np.random.normal(loc=50, scale=10, size=30)
group2 = np.random.normal(loc=55, scale=10, size=30)
group3 = np.random.normal(loc=60, scale=10, size=30)

# Check assumptions before performing ANOVA
check_anova_assumptions([group1, group2, group3])

# Visualization of group distributions
plt.figure(figsize=(8, 6))
sns.violinplot(data=[group1, group2, group3])
plt.xticks(ticks=[0, 1, 2], labels=["Group 1", "Group 2", "Group 3"])
plt.title("Violin Plot of Three Groups")
plt.ylabel("Values")
plt.grid()
plt.show()
```

```python
#Q24) Perform a hypothesis test for population variance using a Chi-square distribution and interpret the results
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Simulate random data from a normal distribution
np.random.seed(42)
sample = np.random.normal(loc=50, scale=10, size=30)

# Population variance assumption
population_variance = 100  # Assume known variance for testing

# Calculate sample variance
sample_variance = np.var(sample, ddof=1)

# Perform Chi-square test for variance
n = len(sample)
chi_square_stat = (n - 1) * sample_variance / population_variance
p_value = 1 - stats.chi2.cdf(chi_square_stat, df=n-1)  # One-tailed test

# Print results
print(f"Chi-square Statistic: {chi_square_stat:.4f}")
print(f"P-value: {p_value:.4f}")
if p_value < 0.05:
    print("The population variance significantly differs from the assumed variance.")
else:
    print("There is no significant difference in variance.")

# Visualize the Chi-square distribution
x = np.linspace(0, 50, 1000)
y = stats.chi2.pdf(x, df=n-1)

plt.figure(figsize=(8, 6))
plt.plot(x, y, label=f'Chi-square Distribution (df={n-1})', color='blue')
plt.axvline(chi_square_stat, color='red', linestyle='dashed', linewidth=2, label='Observed Chi-square')
plt.fill_between(x, y, where=(x >= chi_square_stat), color='red', alpha=0.3)
plt.title("Chi-square Distribution and Test for Variance")
plt.xlabel("Chi-square Value")
plt.ylabel("Density")
plt.legend()
plt.grid()
plt.show()
```

```python
# Q25)  Write a Python script to perform a Z-test for comparing proportions between two datasets or groups
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Simulate data for two groups (successes and trials)
np.random.seed(42)
successes_group1 = np.random.binomial(n=100, p=0.6, size=1)[0]
successes_group2 = np.random.binomial(n=100, p=0.5, size=1)[0]
trials_group1 = 100
trials_group2 = 100

# Compute proportions
prop1 = successes_group1 / trials_group1
prop2 = successes_group2 / trials_group2

# Compute pooled proportion
pooled_prop = (successes_group1 + successes_group2) / (trials_group1 + trials_group2)

# Compute standard error
se = np.sqrt(pooled_prop * (1 - pooled_prop) * (1/trials_group1 + 1/trials_group2))

# Compute Z-score
z_score = (prop1 - prop2) / se

# Compute p-value (two-tailed test)
p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))

# Print results
print(f"Proportion 1: {prop1:.4f}, Proportion 2: {prop2:.4f}")
print(f"Z-score: {z_score:.4f}")
print(f"P-value: {p_value:.4f}")
if p_value < 0.05:
    print("There is a significant difference between the two proportions.")
else:
    print("No significant difference between the two proportions.")

# Visualization
x = np.linspace(-4, 4, 1000)
y = stats.norm.pdf(x)

plt.figure(figsize=(8, 6))
plt.plot(x, y, label='Standard Normal Distribution', color='blue')
plt.axvline(z_score, color='red', linestyle='dashed', linewidth=2, label='Observed Z-score')
plt.fill_between(x, y, where=(x >= abs(z_score)), color='red', alpha=0.3)
plt.fill_between(x, y, where=(x <= -abs(z_score)), color='red', alpha=0.3)
plt.title("Z-test for Comparing Proportions")
plt.xlabel("Z-score")
plt.ylabel("Density")
plt.legend()
plt.grid()
plt.show()
```

```python
# Q26) Implement an F-test for comparing the variances of two datasets, then interpret and visualize the resultsD
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Simulate two datasets
np.random.seed(42)
data1 = np.random.normal(loc=50, scale=10, size=30)  # Dataset 1
data2 = np.random.normal(loc=52, scale=15, size=30)  # Dataset 2

# Compute variances
var1 = np.var(data1, ddof=1)
var2 = np.var(data2, ddof=1)

# Perform F-test
F_statistic = var1 / var2
p_value = stats.f.cdf(F_statistic, dfn=len(data1)-1, dfd=len(data2)-1) * 2  # Two-tailed test

# Print results
print(f"Variance of dataset 1: {var1:.4f}")
print(f"Variance of dataset 2: {var2:.4f}")
print(f"F-statistic: {F_statistic:.4f}")
print(f"P-value: {p_value:.4f}")
if p_value < 0.05:
    print("Significant difference in variances.")
else:
    print("No significant difference in variances.")

# Visualization
x = np.linspace(0, 5, 1000)
y = stats.f.pdf(x, dfn=len(data1)-1, dfd=len(data2)-1)

plt.figure(figsize=(8, 6))
plt.plot(x, y, label='F-distribution', color='blue')
plt.axvline(F_statistic, color='red', linestyle='dashed', linewidth=2, label='Observed F-statistic')
plt.title("F-test for Comparing Variances")
plt.xlabel("F-statistic")
plt.ylabel("Density")
plt.legend()
plt.grid()
plt.show()
```

```python
# Q27) D Perform a Chi-square test for goodness of fit with simulated data and analyze the results.
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Simulate observed data (categorical counts)
np.random.seed(42)
observed = np.array([50, 30, 20])  # Example observed counts
expected = np.array([40, 40, 20])  # Expected proportions based on hypothesis

# Perform Chi-square goodness-of-fit test
chi2_stat, p_value = stats.chisquare(f_obs=observed, f_exp=expected)

# Print results
print(f"Chi-square Statistic: {chi2_stat:.4f}")
print(f"P-value: {p_value:.4f}")
if p_value < 0.05:
    print("The observed distribution significantly differs from the expected distribution.")
else:
    print("No significant difference between observed and expected distributions.")

# Visualization
categories = ['Category 1', 'Category 2', 'Category 3']
bar_width = 0.4
indices = np.arange(len(categories))

plt.figure(figsize=(8, 6))
plt.bar(indices, observed, bar_width, label='Observed', color='blue', alpha=0.7)
plt.bar(indices + bar_width, expected, bar_width, label='Expected', color='orange', alpha=0.7)
plt.xticks(indices + bar_width / 2, categories)
plt.ylabel("Counts")
plt.title("Chi-square Goodness-of-Fit Test")
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
